```python
data = pd.read_csv(r'\Users\chris\Downloads\wwq.csv')
    
df = (   
    pd.DataFrame(data)
    .rename(columns={"alcohol": "alc"}) #Abbreviating longer column names
    .rename(columns={"free sulfur dioxide": "F.S.D."})
    .rename(columns={"total sulfur dioxide": "T.S.D."})
    .sort_values("quality", ascending=False) #Sorting by wine Quality to answer our research question
    .reset_index(drop=True)
)

df.head(10)
```

```python
#Our data set is already clean with no missing values, so our function just cleans up column names, and sorts by wine quality.

def load_and_process(path): 
    
    data = pd.read_csv(path)
    
    newdf = (   
    pd.DataFrame(data)
    .rename(columns={"alcohol": "Alc"}) #Abbreviating longer column names
    .rename(columns={"free sulfur dioxide": "F.S.D."})
    .rename(columns={"total sulfur dioxide": "T.S.D."})
    .sort_values("quality", ascending=False) #Sorting by wine Quality to answer our research question
    .reset_index(drop=True) #Adjusted Index to match our new sorted dataframe
    )

    return newdf
```

```python
df3 = load_and_process(r'\Users\chris\Downloads\wwq.csv')
df3.head(15)
```

```python
# Returns expected output, the functions retitles columns and sorts by the wine quality, 
# making it easier for us to further manipulate the df to conduct our analysis
```
